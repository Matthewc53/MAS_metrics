```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
csv = pd.read_csv('../Doc_Type_Frequency_data.csv')

df = pd.DataFrame(csv)

df['Form Name'] = df['Form Name'].str.strip()

dfcor = df[df['Form Name'].isin(['Correspondence', 'Third Party Correspondence'])]

dfret = df.loc[df['Form Name'].str.contains('21-22|21-4138|21-0966|Returned Mail')]

df['Date Received'] = pd.to_datetime(df['Date Received'])

df['Form Name'] = df['Form Name'].replace(to_replace='VA 21-22 Appointment of Veterans Serv. Org. as Cla', value ='Power of Attorney incl. VA 21-22, VA 22a', regex=True)

df
```

```python
merged = pd.concat([dfcor, dfret]) 


merged = merged.groupby(['Date Received','Form Name']).size()


merged = pd.DataFrame(merged)
```

```python


Ret_Mail = merged.loc[(slice(None), 'Returned Mail'), :]

Correspondence = merged.loc[(slice(None), 'Correspondence'), :]

to_on = merged.loc[(slice(None), 'VA 21-22 Appointment of Veterans Serv. Org. as Cla'), :]

evi = merged.loc[(slice(None), 'VA 21-4138 Statement In Support of Claim'), :]

tpc = merged.loc[(slice(None), 'Third Party Correspondence'), :]

nin_sx = tpc = merged.loc[(slice(None), 'VA 21-0966 Intent to File'), :]




```

```python
Ret_Mail_ind = Ret_Mail.index.tolist()
```

```python
Ret_Mail_ind = [('5/1/2020'),
 ('5/2/2020'),
 ('5/3/2020'),
 ('5/4/2020'),
 ('5/5/2020'),
 ('5/6/2020'),
 ('5/7/2020')]
```

```python
plt.figure(figsize=[20,20])

plt.plot(Ret_Mail_ind, Ret_Mail[0], linewidth=10, label='Returned Mail')

plt.plot(Ret_Mail_ind, Correspondence[0], linewidth=10, label='Correspondence')

plt.plot(Ret_Mail_ind, to_on[0], linewidth=10, label='21-22')

plt.plot(Ret_Mail_ind, evi[0], linewidth=10, label='4138')

plt.plot(Ret_Mail_ind, tpc[0], linewidth=10, label='TPC')

plt.plot(Ret_Mail_ind, tpc[0], linewidth=10, label='0966')

plt.ylabel('Number of Daily documents', fontsize=20)

plt.yticks(fontsize=25)

plt.xticks(fontsize=25)

plt.legend(fontsize=15)

plt.title('Daily document counts(Doctype Frequency Tableau)', fontsize=20)

plt.savefig('Daily document counts(Doctype Frequency Tableau).png')
```
